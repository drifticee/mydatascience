```python
city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']

#위도와 경도
lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

#인구 밀도
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den) * 0.2
#마커의 크기를 인구 밀도에 따라 정함
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y']

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5)
#여기서 alpha는 투명도 의미
plt.xlabel('경도(longitude)')
plt.ylabel('위도(latitude)')
plt.title('지역별 인구 밀도(2017)')

for x, y, name in zip(lon, lat, city):
    plt.text(x, y, name)
    #위도와 경도에 맞게 도시 이름 출력

plt.show()
```

```python
#막대 그래프 (bar graph)

member_IDs = ['m_01', 'm_02', 'm_03', 'm_04']
before_ex = [27, 35, 40, 33]
after_ex = [30, 38, 42, 37]
colors = ['r', 'g', 'b', 'm']

n_data = len(member_IDs)
index = np.arange(n_data)
plt.bar(index, before_ex, color = colors, tick_label = member_IDs, width = 0.6)
#x = index, y = before_ex,
#막대 별 컬러 설정,
#x축을 숫자 대신 각 막대 그래프 라벨로 변경,
#막대의 폭 변경


plt.show()
```

```python
#여러 엑셀 파일 데이터 수정
import re

excel_data_files1 = glob.glob('C:\\users\\student\\mypy\\data\\담당자별_판매량_*사원.xlsx')

for f in excel_data_files1:
    df = pd.read_excel(f)
    
    #데이터 변경
    if(df.loc[1, '담당자'] == 'A'):
        df['담당자'] = 'Andy'
    elif(df.loc[1, '담당자'] == 'B'):
        df['담당자'] = 'Becky'
    elif(df.loc[1, '담당자'] == 'C'):
        df['담당자'] = 'Chris'
    #파일 전체의 담당자 열 데이터를 수정하는 방식
    #파일 속 담당자가 여러 명이면 각각 if문으로 체크하고 하나씩 수정하는 게 맞을 듯
    #파일 하나엔 담당자 한 명이라 가정하고 하는 것이라 elif도 필요없음 if만 써도 된다
    
    f_new = re.sub('.xlsx', '2.xlsx', f)
    #문자열 처리 모듈인 re 사용하여 파일명 변경
    
    new_excel_writer = pd.ExcelWriter(f_new, engine='xlsxwriter')
    df.to_excel(new_excel_writer, index=False)
    new_excel_writer.save()
```

```python
#위의 차트는 범례도 default로 나와있고 x축 값도 기본인 1부터 시작함
#add_series(~, 'categories'= x축의 데이터값 지정, 'name'= 데이터 이름 지정(범례))
excel_chart = pd.ExcelWriter('C:\\users\\student\\mypy\\data\\data_chart_in_excel.xlsx', engine='xlsxwriter')

df.to_excel(excel_chart, index=True, sheet_name='Sheet1')

worksheet = excel_chart.sheets['Sheet1']

chart = excel_chart.book.add_chart({'type': 'line'})

chart.add_series({'values': '=Sheet1!$B$2:$B$8',
                 'categories': '=Sheet1!$A$2:$A$8',
                 'name': '=Sheet1!$B$1'})
chart.add_series({'values': '=Sheet1!$C$2:$C$8',
                 'categories': '=Sheet1!$A$2:$A$8',
                 'name': '=Sheet1!$C$1'})

worksheet.insert_chart('D2', chart)

excel_chart.save()
```
