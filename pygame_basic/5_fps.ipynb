```python
import pygame

pygame.init() #pygame은 초기화 필요

#화면 크기 설정
screen_width = 480
screen_height = 640
screen = pygame.display.set_mode((screen_width, screen_height))

#화면 타이틀
pygame.display.set_caption('Nado Game')

#FPS
clock = pygame.time.Clock()

#배경 이미지 불러오기
background = pygame.image.load('C:\\users\\Yubin\\mypy\\pygame_basic\\background.png')

#캐릭터(스프라이트) 불러오기
character = pygame.image.load('C:\\users\\Yubin\\mypy\\pygame_basic\\character.png')
character_size = character.get_rect().size #이미지의 크기를 구해옴
character_width = character_size[0]
character_height = character_size[1]
character_x_pos = (screen_width - character_width)/2 #화면 가로 중앙
character_y_pos = screen_height - character_height   #바닥 위치

#이동할 좌표
to_x = 0
to_y = 0

char_x_speed = 0.5
char_y_speed = 0.5

#이벤트 루프
running = True
while running: #게임 진행 동안
    dt = clock.tick(75) #게임 화면의 초당 프레임 수 설정
    #우리가 보는 이동 속도 = 프레임 x 설정한 속도 - 프레임 별로 속도 맞춰줌
    
    for event in pygame.event.get(): #어떤 이벤트가 발생?
        if event.type == pygame.QUIT: #게임 창을 닫으면 running 에 False를 줌
            running = False
            
        if event.type == pygame.KEYDOWN: #키보드를 눌렀을 때
            if event.key == pygame.K_LEFT:
                to_x -= char_x_speed
            elif event.key == pygame.K_RIGHT:
                to_x += char_x_speed
            elif event.key == pygame.K_UP:
                to_y -= char_y_speed
            elif event.key == pygame.K_DOWN:
                to_y += char_y_speed
                
        if event.type == pygame.KEYUP: #방향키를 뗐을 때
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                to_x = 0
            elif event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                to_y = 0
    
    character_x_pos += to_x * dt
    character_y_pos += to_y * dt
    
    if character_x_pos < 0:
        character_x_pos = 0
    elif character_x_pos > screen_width - character_width:
        character_x_pos = screen_width - character_width
    elif character_y_pos < 0:
        character_y_pos = 0
    elif character_y_pos > screen_height - character_height:
        character_y_pos = screen_height - character_height
    
    screen.blit(background, (0,0)) #배경 그리기
    screen.blit(character, (character_x_pos, character_y_pos))
    pygame.display.update() #게임화면 다시 그리기!

pygame.quit()
```
