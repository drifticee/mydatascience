```python
import rdflib
from rdflib import Graph

g = Graph()
path = 'C:\\Users\\Yubin\\law_data\\'
g.parse(path + 'nia_law_ontology_판례_20180117.nt', format = 'nt')
```

```python
import numpy as np
from konlpy.tag import Okt
from sklearn.feature_extraction.text import CountVectorizer

keyword = '배상'

okt = Okt()
count = CountVectorizer()

c=0
for st in g:
    c+=1
    if c>200:
        break
        
    if keyword in okt.nouns(st[2]):
        print(okt.morphs(st[2]))
        bag_of_words = count.fit_transform(np.array([st[2]]))
        print(bag_of_words.toarray()) #이를 행렬로
        print(count.get_feature_names()) #특성 이름들
```
